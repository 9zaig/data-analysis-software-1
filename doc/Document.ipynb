## III / Relations between the data

```python
nb_testSite = df['TestSite'].value_counts().values
labels = ['DCM_LMZC','DCM_LMZO']
```

### WAFER ID VS SOME NUMERICAL FEATURES

do the wafer id influence the data ?

```python
plt.figure(figsize=(12,8))
ans = plt.subplot(2,2,1)
plt.title('I at -1V [A]')
plt.scatter(df['Wafer'], df['I at -1V [A]'])
ans = plt.subplot(2,2,2)
plt.title('I at 1V [A]')
plt.scatter(df['Wafer'], df['I at 1V [A]'])
ans = plt.subplot(2,2,3)
plt.title('Rsq of IV (1)')
plt.scatter(df['Wafer'], df['Rsq of IV (1)'])
ans = plt.subplot(2,2,4)
plt.title('Max Transmission of Ref.spec.(dB)')
plt.scatter(df['Wafer'], df['Max Transmission of Ref.spec.(dB)'])
plt.show()
```

We can see that the current at 1V are really dependent of the wafer id. We can also see a few outliers, but mostly the current at 1V from wafer D07 and D08 are closer to 0.5A whereas the current at 1V from the wafer D23 are closer to 2A and for the wafer D24 it's closer to 2.5A or 3A

### TESTSITE VS SOME NUMERICAL FEATURES

do the test site influence the data ?

```python
plt.figure(figsize=(12,8))
ans = plt.subplot(2,2,1)
plt.title('I at -1V [A]')
plt.scatter(df['TestSite'], df['I at -1V [A]'])
ans = plt.subplot(2,2,2)
plt.title('I at 1V [A]')
plt.scatter(df['TestSite'], df['I at 1V [A]'])
ans = plt.subplot(2,2,3)
plt.title('Rsq of IV (1)')
plt.scatter(df['TestSite'], df['Rsq of IV (1)'])
ans = plt.subplot(2,2,4)
plt.title('Max Transmission of Ref.spec.(dB)')
plt.scatter(df['TestSite'], df['Max Transmission of Ref.spec.(dB)'])
plt.show()
```

we can see that for the max Transmission of reference spectrum there is a clear difference between the LMZC test Site and the LMZO test Site. For the others graph we can't really say that there are clear difference between the test Site, what we can say is that the test Site LMZO seems to be more scatter than the test site LMZC.
